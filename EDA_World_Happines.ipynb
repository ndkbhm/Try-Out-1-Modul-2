```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
whr = pd.read_csv('2019.csv')
whr
```

```python
# Melihat kurva dan melihat tipe distribusi data yang condong miring ke kanan
sns.set(color_codes=True)
sns.set_style("whitegrid")
var = ['Generosity', 'Perceptions of corruption']
sns.displot(whr[var], kde="reg")
```

```python
# Menghitung Central Tendency menggunakan median(), 
# karena lebih cocok digunakan untuk distribusi miring
whr_central_tendency = whr[['Generosity','Perceptions of corruption']].median()
whr_central_tendency
```

```python
# Menggunakan perhitungan standar deviasi
whr_dispersion = whr[['Generosity','Perceptions of corruption']].std()
whr_dispersion
```

```python
# Menampilkan outlier menggunakan boxplot
sns.boxplot(x=whr['Social support'])
```

```python
sns.boxplot(x=whr['Freedom to make life choices'])
```

```python
whr_corr = whr.groupby('Score')[['GDP per capita', 'Social support', 'Healthy life expectancy']].corr()
whr_corr
```
